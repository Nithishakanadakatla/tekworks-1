```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Practice-Datasets .csv")
df
```

```python
df["Purchase_Date"] = pd.to_datetime(df["Purchase_Date"])
```

```python
# Histogram to analyze Age distribution
plt.figure()
plt.hist(df["Age"], bins=15, color="skyblue")
plt.title("Age Distribution")
plt.show()

```

```python
# Pie chart showing Gender distribution
plt.figure()
df["Gender"].value_counts().plot.pie(autopct="%1.1f%%")
plt.ylabel("")
plt.show()
```

```python
#  Number of customers in each city
plt.figure()
df["City"].value_counts().plot.bar(color="red")
plt.show()
```

```python
#  Box plot for Monthly Income distribution
plt.figure()
plt.boxplot(df["Monthly_Income"], patch_artist=True,
            boxprops=dict(facecolor="lightgreen"))
plt.show()

```

```python
#  Histogram of Spending Score
plt.figure()
plt.hist(df["Spending_Score"], bins=15)
plt.show()

# skewness value
print("Skewness:", df["Spending_Score"].skew())
```

```python
#  Purchase Amount trend over time
trend = df.groupby("Purchase_Date")["Purchase_Amount"].sum()
plt.figure()
plt.plot(trend, color="purple")
plt.show()
```

```python
#  Average Monthly Income by Gender
plt.figure()
df.groupby("Gender")["Monthly_Income"].mean().plot.bar(color=["blue","pink"])
plt.show()

```

```python
# Purchase Amount grouped by Product Category
plt.figure()
df.boxplot(column="Purchase_Amount", by="Product_Category",
           patch_artist=True,
           boxprops=dict(facecolor="cyan"))
plt.suptitle("")
plt.show()
```

```python
#  Product Category percentage
plt.figure()
df["Product_Category"].value_counts().plot.pie(autopct="%1.1f%%")
plt.ylabel("")
plt.show()
```

```python
# Histogram of Purchase Amount
plt.figure()
plt.hist(df["Purchase_Amount"], bins=20, color="brown")
plt.show()
```

```python
#  Average Spending Score by City
plt.figure()
df.groupby("City")["Spending_Score"].mean().plot.bar(color="teal")
plt.show()
```

```python
#  Monthly total purchase amount
monthly = df.resample('M', on='Purchase_Date')["Purchase_Amount"].sum()
plt.figure()
plt.plot(monthly, color="darkgreen")
plt.show()
```

```python
#  Monthly Income comparison across cities
plt.figure()
df.boxplot(column="Monthly_Income", by="City",
           patch_artist=True,
           boxprops=dict(facecolor="lightyellow"))
plt.suptitle("")
plt.show()
```

```python
#  Age distribution by gender
plt.figure()
for g in df["Gender"].unique():
    plt.hist(df[df["Gender"]==g]["Age"], bins=15, alpha=0.5, label=g)

plt.legend()
plt.show()
```

```python
#  Total purchase by product category
plt.figure()
df.groupby("Product_Category")["Purchase_Amount"].sum().plot.bar(color="magenta")
plt.show()
```

```python
#  Relationship between Monthly Income and Spending Score
plt.figure()
plt.scatter(df["Monthly_Income"], df["Spending_Score"], color="lavender")
plt.show()
```

```python
# Daily purchase amount variation
daily = df.groupby(df["Purchase_Date"].dt.date)["Purchase_Amount"].sum()
plt.figure()
plt.plot(daily, color="navy")
plt.show()
```

```python
#  Spending Score distribution by gender
plt.figure()
df.boxplot(column="Spending_Score", by="Gender",
           patch_artist=True,
           boxprops=dict(facecolor="lightpink"))
plt.suptitle("")
plt.show()

```

```python
#  Customer count by product category and gender
plt.figure()
pd.crosstab(df["Product_Category"], df["Gender"]).plot.bar()
plt.show()
```

```python
# Relationship between Age and Purchase Amount
plt.figure()
plt.scatter(df["Age"], df["Purchase_Amount"],color="pink",edgecolor="red")
plt.show()
```
